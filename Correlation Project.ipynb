## Correlation Project

```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)  # used to adjust sixe of plots 

# Reading data
df= pd.read_csv(r"C:\Users\vishn\Downloads\archive (5)\movies.csv")
```

```python
df.head()
```

### Data Cleaning

```python
# Checking for any missing data
for col in df.columns:
    missing_data=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,missing_data))
```

```python
# Removing the missing data
df=df.dropna()
```

```python
df.dtypes
```

```python
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df[['yearcorrect','released']].head()
```

```python
# Ordering the data
df.sort_values(by=['gross'],inplace=False,ascending=False).head()
```

### Finding Correlations in the Data

 **Predictions**:
 * High Correlation with Budget 
 * High correlation with Company

#### Budget

```python
# Scatter Plot : Budget vs Gross Revenue
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget and Gross Revenue')
plt.xlabel('Gross Revenue')
plt.ylabel('Budget')


plt.show()
```

```python
# regression plot for budget vs gross
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
# Looking at correlation
df.corr(method='pearson')
```

```python
corr_matrix=df.corr(method='pearson')
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

* The correlation of Budget and Gross revenue is 0.74, which lies on the higher range and thus proves our predicition as correct

#### Company

```python
# Since company is not numeric, we can turn it into a numeric value
df.head()
```

```python
df_numeric = df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype=='object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes

df_numeric

       
```

```python
corr_matrix=df_numeric.corr(method='pearson')
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

* The company's correlation with gross revenue is 0.15, indicating a very low correlation

```python
df_numeric.corr()
```

```python
# Unstacking the data
corr_matrix=df_numeric.corr()
corr_pairs=corr_matrix.unstack()
corr_pairs
```

```python
sorted_corr=corr_pairs.sort_values()
```

```python
sorted_corr
```

```python
# Checking all the correlations more than 0.5
high_corr=sorted_corr[(sorted_corr)>0.5]

high_corr
```

* Votes and budget revenue have the highest correlation ot gross revenue
